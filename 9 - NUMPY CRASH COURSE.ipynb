```python
# variance of array
s = np.array([1,2,3,4])
var = np.var(s)
print('Variance of S : ' , var)
```

```python
# standard deviation of array
s = np.array([1,2,3,4])
std_dev = np.std(s)
print('Standard Deviation of s :' , std_dev)
```

# Boolean & Logical Function

```python
# check if all elements are true
# All
logical_test = np.array([False,True,True])
all_true = np.all(logical_test) # check if all elemmts are true
print('All elements are true :' , all_true)
```

```python
# check if all elements are true

logical_test = np.array([False,True,False])
all_true = np.all(logical_test) # check if all elemmts are true
print('All elements are true :' , all_true)
```

```python
# check if all elements are true

logical_test = np.array([True,True,False])
all_true = np.all(logical_test) # check if all elemmts are true
print('All elements are true :' , all_true)
```

```python
#any
any_true = np.any(logical_test)
print('any elements are true :',any_true)
```

```python
logical_test = np.array([False,False,False])
any_true = np.any(logical_test)
print('any elements are true :',any_true)
```

# Set operations

```python
# Intersection of two arrays
set_a = np.array([1, 2, 3, 4])
set_b = np.array([3, 4, 5, 6])
intersection = np.intersect1d(set_a, set_b)
print("Intersection of a and b:", intersection)
```

```python
#Union 
union = np.union1d(set_a , set_b)
print('Union of a and b :' , union)
```

# Array Atribute Function

```python
# array attribute
a = np.array([1,2,3])
shape = a.shape # Shape of array
size = a.size # number of elements
dimensions = a.ndim # number of dimenstion
dtype = a.dtype # Data type of array

print('Shape of array :' , shape)
print('Size of array:' , size)
print('Dimentions of array:',dimensions)
print('datatype of array : ' , dtype)
```
